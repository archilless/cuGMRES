```python
residual_means_N = []
residual_stds_N = []

prefix = "/media/linux/4db3d51d-3503-451d-aff7-07e3ce95927e/Archive/Output/Relative_errors/time_"
for power in tqdm_notebook(range(8, 13)):
    residual_means = []
    residual_stds = []
    N = 2**power
    for maxiter in range(3, 50):
        residual_test = []
        for repetition in range(100):
            residual_test.append(get_float_value(prefix + str(N) 
                                + "/relative_error_" + str(maxiter * 100 + repetition) +  ".txt"))
        append_refined_statistics(residual_test, residual_means, residual_stds)
    residual_means_N.append(residual_means)
    residual_stds_N.append(residual_stds)
    
for power in tqdm_notebook([13]):
    residual_means = []
    residual_stds = []
    N = 2**power
    for maxiter in range(3, 40):
        residual_test = []
        for repetition in range(100):
            residual_test.append(get_float_value(prefix + str(N) 
                                + "/relative_error_" + str(maxiter * 100 + repetition) +  ".txt"))
        append_refined_statistics(residual_test, residual_means, residual_stds)
    residual_means_N.append(residual_means)
    residual_stds_N.append(residual_stds)
```
